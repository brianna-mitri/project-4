## Setup

```python
# 1. Import Libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x_test = pd.read_csv("Output/Split_Train_Test/dest/x_test.zip")
x_train = pd.read_csv("Output/Split_Train_Test/dest/x_train.zip")
y_test = pd.read_csv("Output/Split_Train_Test/dest/y_test.zip")
y_train = pd.read_csv("Output/Split_Train_Test/dest/y_train.zip")


display(x_test.head())
display(x_train.head())
display(y_test.head())
display(y_train.head())
```

```python
x_train.columns
```

```python
#  Train Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(x_train, y_train.values.ravel())
```

```python
clf.score(x_train, y_train),clf.score(x_test, y_test)
```

```python
#  Predictions & Evaluation
y_pred = clf.predict(x_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
#  Calculate Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# 2. Create DataFrame
cm_df = pd.DataFrame(
    cm,
    index=[f"Actual {cls}" for cls in clf.classes_],
    columns=[f"Predicted {cls}" for cls in clf.classes_]
)

# 3. Calculate Accuracy Score
acc_score = accuracy_score(y_test, y_pred)
```

```python
# 4. Displaying Results
print("Confusion Matrix:")
display(cm_df)
print(f"Accuracy Score: {acc_score:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
#  Plot Heatmap (Optional but cleaner!)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix Heatmap')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
importances = pd.Series(clf.feature_importances_, index=x_train.columns).sort_values(ascending=False)
importances_df = pd.DataFrame({
    'Feature': importances.index,
    'Importance Score': importances.values
})
print("Feature Importances:")
display(importances_df.head(5))
```

```python
importances = pd.Series(clf.feature_importances_, index=x_train.columns).sort_values(ascending=False)
importances_df = pd.DataFrame({
    'Feature': importances.index,
    'Importance Score': importances.values
})
print("Feature Importances:")
display(importances_df.head(5))
```

```python
# Extract top 20 features
importances_df_top10 = importances_df.head(10)
```

```python
# Visualize the features by importance
plt.figure(figsize=(10, 8))
plt.barh(importances_df_top10['Feature'], importances_df_top10['Importance Score'])
plt.xlabel('Importance Score')
plt.title('Top 20 Feature Importances')
plt.gca().invert_yaxis()  # Ensures the most important features are at the top
plt.tight_layout()
plt.show()
```
