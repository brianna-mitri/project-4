## Setup

```python
# 1. Import Libraries
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
folder = '../../Preprocessing/Output/Split_Train_Test/'

x_test = pd.read_csv(f"{folder}x_test.zip")
x_train = pd.read_csv(f"{folder}x_train.zip")
y_test = pd.read_csv(f"{folder}y_test.zip")
y_train = pd.read_csv(f"{folder}y_train.zip")

display(x_test.head())
display(x_train.head())
display(y_test.head())
display(y_train.head())
```

```python
#  Train K Nearest Neighbors Classifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(x_train, y_train.values.ravel())
```

```python
clf.score(x_train, y_train),clf.score(x_test, y_test)
```

```python
#  Predictions & Evaluation
y_pred = clf.predict(x_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
#  Calculate Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# 2. Create DataFrame
cm_df = pd.DataFrame(
    cm,
    index=[f"Actual {cls}" for cls in clf.classes_],
    columns=[f"Predicted {cls}" for cls in clf.classes_]
)

# 3. Calculate Accuracy Score
acc_score = accuracy_score(y_test, y_pred)
```

```python
# 4. Displaying Results
print("Confusion Matrix:")
display(cm_df)
print(f"Accuracy Score: {acc_score:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
#  Plot Heatmap (Optional but cleaner!)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix Heatmap')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
